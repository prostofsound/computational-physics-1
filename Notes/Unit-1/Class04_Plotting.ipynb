# Plotting in Python

Matplotlib is a library used for visualization with Python. It allows users to create static, animated, and interactive plots.

"It makes easy things easy and hard things possible." (from Matplotlib website) Customising plots is easy. The documentation is comprehensive and easy to find.

pyplot():
Basic plotting with pyplot

Pyplot is an interface to plot using Matlab- and Mathematica-like plotting sequences. It is a function within matplotlib.

It facilitates plotting. Plotting can be done without using "objects", e.g.

```python
# Import third-party libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a couple of vectors
x = np.arange(0, 6*np.pi, 0.01)
y = np.sin(x)

# Plotting
plt.plot(x,y)
```

```python
# It is recommended to use figure environments instead.

# Open env.
plt.figure()

# Plot
plt.plot(x,y)

# Show/Save
plt.show()
# plt.savefig() #place this line before showing the plot, to avoid closing it and saving nothing

# In actual code, we should always clean up the memory
# plt.close() #close after saving the image, to not run out of memory.
```

```python
# Customise our figures:

# Open env and stablish the figure size
plt.figure(figsize=(4,3))

# Plot
plt.plot(x, y, color = "red", linestyle = ":") #specify the color and linestyle

# Title with latex syntax
plt.title(r"$\sin(x)$") #the r is necessary for the function to recognize this format

# Axes titles
plt.xlabel(r"$x$")
plt.ylabel(r"$y=\sin(x)$")

# Show/Save
plt.show()
plt.close()
```

## Overplotting

```python
# Superimpose plots:

w = np.cos(x)

# Open env.
plt.figure(figsize=(4,3))

# Plot
plt.plot(x, y, color = "red", linestyle = ":", label = r"$\sin(x)$")
plt.plot(x, w, color = "green", linestyle = "-.", label = r"$\cos(x)$")

# Title with latex syntax
plt.title(r"$\sin(x)$ and $\cos(x)$")

# Axes titles
plt.xlabel(r"$x$")
plt.ylabel(r"$y,w$")

# Legend is added with loc for location
plt.legend(loc=4) # loc -> LOCATION

# Show/Save
plt.savefig("./sincos.eps",format = 'eps')
plt.show()
```

## Multi-panel figures

```python
# Method 1: Multi-panel figure

plt.figure()

plt.subplot(2, 2, 1) #first two elements size of panel, the third is to specify the subplot location on this panel.
plt.plot(x, x, color = "red", label = "linear")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.subplot(2, 2, 2)
plt.plot(x, x**2, color = "magenta", label = "parabolic")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.subplot(2, 2, 3)
plt.plot(x, x**3, color = "blue", label = "3rd order")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.subplot(2, 2, 4)
plt.plot(x, x**4, color = "green", label = "4th order")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.show()
plt.close()
```

```python
# Method 2: Multi-panel figure using axes objects 

fig, ax = plt.subplots(nrows = 2, ncols = 2) #this is a tuple, the figure object is the panel and ax is a (2,2) array
ax1, ax2, ax3, ax4 = ax.flatten() #.flatten() returns a one-dimensional representation of the original array
#in this case we had a 2D array and we 'flattened' it to a 1D array to access the elements and plot using each one
# ax1 = ax[0,0] (top-left)
# ax2 = ax[0,1] (top-right)
# ax3 = ax[1,0] (bottom-left)
# ax4 = ax[1,1] (bottom-right)

ax1.plot(x, x, color = "red", label = "linear")
ax1.legend()
ax2.plot(x, x**2, color = "magenta", label = "parabolic")
ax2.legend()
ax3.plot(x, x**3, color = "blue", label = "3rd order")
ax3.legend()
ax4.plot(x, x**4, color = "green", label = "4th order")
ax4.legend()

plt.show()
plt.close()
```

## Scatter plots

```python
# Let us generate randon numbers

# Set up a seed
np.random.seed(123) #if writing without the seed, we'll get different values each time, this line fixes the seed

# Generates randon numbers
new_x = np.random.randn(50) #randn n is the number of elements to generate
new_y = np.random.randn(50)

print(new_x[-3]) 
print(new_x, new_x.shape, new_y.shape)
```

```python
# Open env.
plt.figure(figsize=(4,3))

# Plot
plt.plot(new_x, new_y, color = "red", linestyle = "", marker = "d", label = r"$rand(x)$") #we have to remove the linestyle, to avoid connecting the points

# Title with latex syntax
plt.title("Random numbers")

# Axes titles
plt.xlabel(r"$rand(x)$")
plt.ylabel(r"$rand(y)$")

# Legend is added with loc for location
plt.legend(loc=3)

# Show/Save
plt.show()
```

## Histograms

```python
# Let's make histograms of new_x and new_y

new_x = np.random.randn(500) #when going to infinity, a poisson distribution converges to a more gaussian distribution
new_y = np.random.randn(500)

#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

count, bins, _ = plt.hist(new_x, bins = 20, alpha = 0.3) #check documentation for the returns
count, bins, _ = plt.hist(new_y, bins = 20, alpha = 0.2) #alpha controls the transparency
```

```python
count, bins, _ = plt.hist(new_x, bins = 20, histtype = "step") #check documentation for the returns
count, bins, _ = plt.hist(new_y, bins = 20, histtype = "step") #alpha controls the transparency
```

```python
# Print the objects that plt.hist() returned

print(count.shape) #count is an array of numbers representing how many data points fall into each bin.
print(bins.shape) #bins is an array specifying the edges (boundaries) of each bin., so we have (bins.shape - 1) bins.
print(count)
```

The data is dividen into bins and it counts how many values fall into each bin.

Bins are the intervals (or "buckets") into which the data is grouped. The number of bins affects the appearance of the histogram:

* Too few bins → the histogram is too coarse, losing details.
* Too many bins → the histogram is too fine, showing too much noise.

```python
# Uniform Distribution

new_z = np.random.uniform(-3,3,500) #lowest value, largest value, numbers to be generated

#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

count1, bins1, unimportant1 = plt.hist(new_z, bins = 20, alpha = 0.3, histtype = "step") #orange one

count2, bins2, unimportant2 = plt.hist(new_x, bins = 20, alpha = 0.7, histtype = "step") #blue one
```

```python
print(count1.shape, bins1.shape)

print(type(count1))
```

```python
plt.figure(figsize=(4,3))

plt.plot(bins1[1:], count1, linestyle=" ", marker = "+", label="uniform")
plt.plot(bins2[1:], count2, linestyle=" ", marker = "d", label="normal")

plt.legend()
plt.show()
```

## Getting the midpoints for the bins

More appropiate than using the endpoints of the bins.

Shifting the arrays.

When fitting a function (like a Gaussian) to histogram data, you often use bin centers as the x-values.
It makes it easier to interpret histograms when superimposing additional data.

```python
print(bins1) #the elements are the edges of the bins
```

```python
print(bins1.shape)
print(bins1[:-1].shape)
print(bins1[:-1])
```

```python
# Shifting the arrays to get the middle values
mid_bins = (bins1[1:] + bins1[:-1])/2

print(mid_bins)
```

```python
plt.figure(figsize=(4,3))

plt.plot(mid_bins, count1, linestyle=" ", marker = "+", label="uniform")
plt.plot(mid_bins, count2, linestyle=" ", marker = "d", label="normal")

plt.legend()
plt.show()
plt.close()
```
