Here you will have access just to a terminal, then your scripts musts be written on a editor, e.g vim.

Is important to know the following conventions for running, writing and executing your files when sharing the computer with other users and for security reasons.

```python
#when using ls -ltr
#r: read, w: write, x: execute
#the first (l: symbolic link, d: directory, -: file): specifies the kind of file.
#the next 3 are reserved for the user (u).
#the next 3 for the group of users who have access to the machine (g).
#everything else: for the world (a).
```

```python
##if at the end, the "world" can write and execute, could be very dangerous. Then, to change permissions this:
## chmod
## chmod a-x script.exe ##-: to quit permission and +: to give permission
#also there exists a number convention. r:4, w:2, x:1
## chmod 777 script.exe ##for give access to everyone
## chmod 750 script.exe ##just for the user to execute it

#what if you change the execute permissions for the user?
#we get a Permission denied message.
```
